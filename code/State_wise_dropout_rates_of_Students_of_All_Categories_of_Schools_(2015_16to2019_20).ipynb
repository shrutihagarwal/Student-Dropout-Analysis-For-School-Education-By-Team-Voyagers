```python
import pandas as pd
df=pd.read_csv(path1)
df.head(5)
```

```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt


# Bar chart using Plotly
fig = px.bar(df, x='State/UT', y=['2015-16 - Elementary - Boys', '2015-16 - Elementary - Girls'],
             title='Elementary School Dropout Rates (2015-16)')
fig.show()

```

```python
import pandas as pd
import plotly.express as px

# Assuming 'df' is your DataFrame

# Filter out states with 0 or NaN values in both 'Boys' and 'Girls' columns
df_filtered = df[(df['2019-20 - Elementary - Boys'] != 0) & (df['2019-20 - Elementary - Girls'] != 0) & (~df['2019-20 - Elementary - Boys'].isna()) & (~df['2019-20 - Elementary - Girls'].isna())]

# Create the grouped bar chart
fig = px.bar(df_filtered, x='State/UT', y=['2019-20 - Elementary - Boys', '2019-20 - Elementary - Girls'],
             title='Total Dropout Rates by Gender Level (2019-20)',
             labels={'variable': 'Category', 'value': 'Dropout Rate'},
             barmode='group')  # Set barmode to 'group' for grouped bars
fig.update_layout(legend_title_text='Gender')
fig.show()


```
