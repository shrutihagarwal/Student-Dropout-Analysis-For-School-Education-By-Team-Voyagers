```python
import plotly.graph_objects as go
import pandas as pd

# Create a DataFrame from the provided data
data = {
    "Year": [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    "Primary Dropout": [4.67, 4.34, 4.13, 6.35, 3.51, 4.45, 1.48, 0.76, 1.45],
    "Upper Primary Dropout": [3.13, 3.77, 4.03, 5.68, 5.02, 4.68, 2.6, 2.27, 3.02],
    "Secondary Dropout": [14.54, 17.86, 17.06, 22.13, 18.93, 17.9, 16.09, 14.04, 12.61],
}

df = pd.DataFrame(data)

# Create a single line chart with trendlines using Plotly
fig = go.Figure()

# Add traces for primary, upper primary, and secondary dropout rates
fig.add_trace(go.Scatter(x=df["Year"], y=df["Primary Dropout"], mode='lines+markers', name='Primary Dropout'))
fig.add_trace(go.Scatter(x=df["Year"], y=df["Upper Primary Dropout"], mode='lines+markers', name='Upper Primary Dropout'))
fig.add_trace(go.Scatter(x=df["Year"], y=df["Secondary Dropout"], mode='lines+markers', name='Secondary Dropout'))

# Add trendlines to the traces
fig.update_traces(line=dict(shape='linear'))

# Customize the layout
fig.update_layout(
    xaxis_title="Year",
    yaxis_title="Dropout Rate",
    title="School Dropout Rates for Primary, Upper Primary & Secondary Students",
)

# Show the chart
fig.show()
```

```python
import plotly.express as px

# Create a DataFrame from the provided data
data = {
    "Year": [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    "Primary Dropout": [4.67, 4.34, 4.13, 6.35, 3.51, 4.45, 1.48, 0.76, 1.45],
    "Upper Primary Dropout": [3.13, 3.77, 4.03, 5.68, 5.02, 4.68, 2.6, 2.27, 3.02],
    "Secondary Dropout": [14.54, 17.86, 17.06, 22.13, 18.93, 17.9, 16.09, 14.04, 12.61],
}

df = pd.DataFrame(data)

# Create a grouped bar chart for primary, upper primary, and secondary dropout rates
fig = px.bar(df, x='Year', y=['Primary Dropout', 'Upper Primary Dropout', 'Secondary Dropout'],
             title='School Dropout Rates Over the Years',
             labels={'value': 'Dropout Rate', 'variable': 'Category'},
             category_orders={'variable': ['Primary Dropout', 'Upper Primary Dropout', 'Secondary Dropout']})

# Customize the layout
fig.update_layout(barmode='group', xaxis_title='Year')

# Show the chart
fig.show()
```
