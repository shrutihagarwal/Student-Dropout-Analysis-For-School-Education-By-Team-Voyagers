```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ["Course Fee", "Books & Stationery", "Transport", "Private Coaching", "Other Expenses"]
urban_male = [57.5, 15.0, 10.2, 12.9, 4.4]
urban_female = [57.3, 15.9, 10.9, 11.5, 4.3]
#urban_overall = [57.4, 15.4, 10.5, 12.3, 4.4]

# Create an array for the x-axis positions
x = np.arange(len(categories))

# Width of each bar
width = 0.25

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(x - width, urban_male, width, label='Male')
plt.bar(x, urban_female, width, label='Female')
#plt.bar(x + width, urban_overall, width, label='Overall')

# Labels, title, and legend
plt.xlabel('Expenditure Categories')
plt.ylabel('Percentage (%)')
plt.title('Urban Expenditure by Gender')
plt.xticks(x, categories)
plt.legend(title='Gender')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
df=pd.read_csv('Average expenditure relating to basic course.csv')
df.head(5)
```

```python
import plotly.express as px

# Assuming you have the data in a pandas DataFrame named 'df'
# If not, create a DataFrame from the given data

# Create a bar plot for rural attendance by education level
fig = px.bar(df, x='level of attendance', y='rural overall', title='Rural Attendance by Education Level')
#fig.update_xaxes(categoryorder='total ascending')
fig.update_xaxes(title='Education Level')
fig.update_yaxes(title='Attendance')
fig.show()

# Create a bar plot for urban attendance by education level
fig = px.bar(df, x='level of attendance', y='urban overall', title='Urban Attendance by Education Level')
fig.update_xaxes(categoryorder='total ascending')
fig.update_xaxes(title='Education Level')
fig.update_yaxes(title='Attendance')
fig.show()
```
