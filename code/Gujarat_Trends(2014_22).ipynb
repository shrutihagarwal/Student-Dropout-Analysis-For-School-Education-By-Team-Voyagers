```python
import plotly.express as px
import pandas as pd

# Create a DataFrame from the provided data
data = {
    "Year": [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    "Primary Dropout": [4.67, 0.76, 4.13, 6.35, 3.51, 4.45, 1.48, 0.76, 1.45],
    "Upper Primary Dropout": [3.13, 5.55, 4.03, 5.68, 5.02, 4.68, 2.6, 2.27, 3.02],
    "Secondary Dropout": [14.54, 21.61, 17.06, 22.13, 18.93, 17.9, 16.09, 14.04, 12.61],
}

df = pd.DataFrame(data)

# Create individual trend charts
fig_primary = px.line(df, x='Year', y='Primary Dropout', title='Primary Dropout Rate Trend')
fig_upper_primary = px.line(df, x='Year', y='Upper Primary Dropout', title='Upper Primary Dropout Rate Trend')
fig_secondary = px.line(df, x='Year', y='Secondary Dropout', title='Secondary Dropout Rate Trend')

# Create a combined trend chart
fig_combined = px.line(df, x='Year', y=['Primary Dropout', 'Upper Primary Dropout', 'Secondary Dropout'],
                       title='Dropout Rate Trends in Gujarat')
fig_combined.update_traces(mode='lines+markers')

# Show the charts
fig_primary.show()
fig_upper_primary.show()
fig_secondary.show()
fig_combined.show()
```

```python
# Create a grouped bar chart for dropout rates
fig_grouped_bar = px.bar(df, x='Year', y=['Primary Dropout', 'Upper Primary Dropout', 'Secondary Dropout'],
                         title='Grouped Dropout Rates in Gujarat')
fig_grouped_bar.update_layout(barmode='group')

# Show the chart
fig_grouped_bar.show()
```

```python
# Calculate the total dropout rates for each category
total_primary_dropout = df['Primary Dropout'].sum()
total_upper_primary_dropout = df['Upper Primary Dropout'].sum()
total_secondary_dropout = df['Secondary Dropout'].sum()

# Create a pie chart for the total dropout rates
data_pie = {
    'Category': ['Primary', 'Upper Primary', 'Secondary'],
    'Total Dropout': [total_primary_dropout, total_upper_primary_dropout, total_secondary_dropout]
}
df_pie = pd.DataFrame(data_pie)

fig_pie = px.pie(df_pie, names='Category', values='Total Dropout',
                 title='Total Dropout Rates in Gujarat by Category')

# Show the chart
fig_pie.show()
```
