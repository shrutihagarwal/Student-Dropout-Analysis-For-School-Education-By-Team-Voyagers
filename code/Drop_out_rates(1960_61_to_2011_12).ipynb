```python
path='/content/drive/MyDrive/DATA HACK/Drop out rates in India from 1960-61 to 2011-12.csv'
```

```python
  import pandas as pd
  data = pd.read_csv(path)
  data.head(5)
```

```python
import plotly.express as px

# Create a line plot using Plotly Express with reduced size
fig = px.line(data, x='Year', y='All Categories - Classes I-X - Total', title='Trend of Students Dropped out in Classes I-X Over Time')
fig.update_xaxes(type='category')
fig.update_layout(xaxis_title='Year', yaxis_title='Dropout', xaxis=dict(tickangle=45), width=800, height=400)
fig.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Create a line plot using Seaborn
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")  # Add a grid background

# Plot the line chart
ax = sns.lineplot(data=data, x='Year', y='All Categories - Classes I-X - Total', marker='o', color='b', label='Dropout Rate')

# Label the dropout values on the lines
for index, row in data.iterrows():
    ax.text(row['Year'], row['All Categories - Classes I-X - Total'], f"{row['All Categories - Classes I-X - Total']:.2f}%",
            ha='left', va='bottom', fontsize=8, color='b')

plt.title('Trend of Students Dropped out in Classes I-X Over Time')
plt.xlabel('Year')
plt.ylabel('Dropout Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

# Create a line plot with separate lines for Boys and Girls
fig = px.line(data, x='Year', y=['All Categories - Classes I-X - Boys', 'All Categories - Classes I-X - Girls'],
              title='Trend of Students Dropped out in Classes I-X Over Time (Boys vs. Girls)')
fig.update_xaxes(type='category')
fig.update_layout(xaxis_title='Year', yaxis_title='Dropout', xaxis=dict(tickangle=45), width=800, height=400)
fig.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Year', y='All Categories - Classes I-X - Total', data=data)
plt.title('Total Dropout in Classes I-V Over Time')
plt.xticks(rotation=45)
plt.show()
```

```python
import plotly.express as px

# Create a grouped bar chart for dropout rates
fig = px.bar(data, x='Year', y=['All Categories - Classes I-X - Boys', 'All Categories - Classes I-X - Girls'],
             title='Dropout Rate Comparison (Boys vs. Girls) Over Time', barmode='group')
fig.update_xaxes(type='category')
fig.update_layout(xaxis_title='Year', yaxis_title='Dropout Rate', xaxis=dict(tickangle=45))
fig.show()
```
